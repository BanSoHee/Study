## Crawling

- crawling_data(url, df, content)

```python
from datetime import datetime, date

## 크롤링 함수 선언
def crawling_data(url, df, content, today, nowtime): #, time):
    
    ## dataframe append data 초기화
    append_value = []
    
    ## crawling url
    driver.get(url)

    ## 검색창 지정
    elem = driver.find_element_by_name("input_search2") # input_search2 = 키워드
    time.sleep(3)
    
    ## 여의도한강1주차장[63빌딩앞] 키워드 검색
    #content = '여의도한강1주차장[63빌딩앞]'
    elem.send_keys(content)
    time.sleep(3)
    elem.send_keys(Keys.RETURN) ## Enter key
    time.sleep(3)

    ## 전체주차면 및 주차가능면 데이터 추출
    html = driver.page_source
    soup = bs(html, 'html.parser')
    search_result = soup.select('.all')

    crawling_data = []
    fin_crawling_data = []

    for i, data in enumerate(search_result):
        #print(i, data.text)
        crawling_data.append(data.text.strip()) ## 텍스트만 추출, 공백 제거
        #print(crawling_data)

    # 전체주차면, 주차가능면 데이터 추출
    for i in crawling_data:
        fin_crawling_data = re.findall(r'\d+', i)

    print(fin_crawling_data)
    
    ## 현재 날짜
    today = today
    
    ## 현재 시간
    nowtime = nowtime
    
    ## dataframe update
    append_value.append(today)
    append_value.append(nowtime)
    append_value.append(content)
    append_value.append(fin_crawling_data[1])
    append_value.append(fin_crawling_data[0])
    
    df.loc[len(df)] = append_value
    
    #print(df)
    return df
```
