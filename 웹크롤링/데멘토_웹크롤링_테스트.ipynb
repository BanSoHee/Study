## Crawling

- crawling_data(url, df, input_time)

```python
## 크롤링 함수 선언
def crawling_data(url, df, input_time): #, time):
    
    ## 서버 날짜 받아오기
    from datetime import datetime, date
    today = date.today()
    today = today.isoformat()
    
    crawling_data = []
    fin_crawling_data = []
    append_value = []
    crawling_time = input_time[3:]
    
    driver.get(url)
    time.sleep(0.5)
    html = driver.page_source

    ## 예) 광나루 - bs4를 이용하여 html parsing
    soup = bs(html, 'html.parser')

    ## 데이터 추출, html 태그 제거, 공백 제거
    search_result = soup.select_one('tbody')
    commu_list = search_result.select('tr > td') ## 데이터 추출

    for i, data in enumerate(commu_list):
        #print(i, data.text)
        crawling_data.append(data.text.strip()) ## 텍스트만 추출, 공백 제거

    # search_result.text
    for i in crawling_data:
        fin_crawling_data.append(i.strip())
        
    ## 길찾기 항목 제거
    remove_set = {'길찾기'}
    fin_crawling_data = [i for i in fin_crawling_data if i not in remove_set]
    #print(fin_crawling_data)
    
    ## 서버 시간 추출
    #search_time = driver.find_element_by_class_name('time-txt')
    #search_time = search_time.text
    
    ## 추출한 데이터를 dataframe에 저장
    append_value.append(today)              ## 크롤링한 서버 시간 추가
    append_value.append(input_time[:2])
    append_value.append(crawling_time)
    
    for i, data in enumerate(fin_crawling_data):    ## 크롤링한 한강 주차 관련 데이터 추가
        #print(i, data)
        
        if len(append_value) == 8:
            #print(append_value)
            df.loc[len(df)] = append_value
            append_value = []
            #print(today, input_time[:2], crawling_time)
            append_value.append(today)
            append_value.append(input_time[:2])
            append_value.append(crawling_time)
            append_value.append(data)
            
        else:
            #print(append_value)
            append_value.append(data)
            
    df.loc[len(df)] = append_value
    
    #print(df)
    
    ## 키보드 입력이 존재할 경우 return
    return df

## 광나루 테스트
#crawling_data(url_Ttukseom, df_Ttukseom)
```
