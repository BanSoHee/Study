```python
from sklearn.datasets import load_iris

## load_iris() API의 반환 결과는 sklearn.utils.Bunch 클래스임.
iris_data = load_iris()
print(type(iris_data))
```

sklearn.datasets 모듈

```python
keys = iris_data.keys() ## 딕셔너리 형태.
print('붓꽃 데이터 세트의 키:', keys)
```

```python
## feature names.
print(type(iris_data.feature_names))
print(len(iris_data.feature_names))
print(iris_data.feature_names)
```

```python
## target names.
print(type(iris_data.target_names))
print(len(iris_data.target_names))
print(iris_data['target_names']) # dict.
print(iris_data.target_names)
```

```python
## data.
print(type(iris_data.data))
print(len(iris_data.data))
print(iris_data.data.shape)
print(iris_data.data)
```

```python
## target = class.
print(type(iris_data.target))
print(len(iris_data.target))
print(iris_data.target.shape)
print(iris_data.target)
```
